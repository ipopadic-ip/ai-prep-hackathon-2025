# Nadgledano uƒçenje

Nadgledano uƒçenje (Supervised Learning) je jedan od najƒçe≈°ƒáe kori≈°ƒáenih pristupa u ma≈°inskom uƒçenju. U ovom pristupu, model se trenira na skupu podataka sa poznatim ulazima (X) i izlazima (y), pri ƒçemu je cilj modela da nauƒçi vezu izmeƒëu ulaznih i izlaznih podataka. Kada se model obuƒçi, mo≈æe se koristiti za predviƒëanje izlaznih vrednosti za nove, nepoznate podatke.

![img/supervised-lr.png](img/supervised-lr.png)


U nadgledanom uƒçenju, model poku≈°ava da nauƒçi funkciju:

ùëì(ùëã)=ùë¶

gde su:

X ‚Äì Ulazni podaci (karakteristike)

y ‚Äì Izlazne vrednosti (cilj)

Podela nadgledanog uƒçenja:
1. Regresija ‚Äì Kada je izlazna vrednost kontinuirana (npr. temperatura, cena, visina).

2. Klasifikacija ‚Äì Kada je izlazna vrednost diskretna (npr. bolest: da/ne, boja: crvena/plava/zuta).

## Regresija


### üåü Analogija
Zamislite da igrate igru gde treba da predvidite koliko ƒáe jabuka biti u korpi na osnovu broja stabala. Ako znate da svako stablo da 5 jabuka, lako je: `jabuke = 5 √ó stabla`. Regresija radi isto ‚Äì pronalazi matematiƒçku vezu izmeƒëu ulaza (stabla) i izlaza (jabuke) da bi predvidela buduƒáe vrednosti.

---

### üìö Definicija
-  Statistiƒçka metoda za **predviƒëanje numeriƒçkih vrednosti** na osnovu istorijskih podataka. Regresija se koristi kada je cilj modela da predvidi kontinualnu vrednost.
- **Primeri**:  
  - Predviƒëanje cene stana na osnovu kvadrature.  
  - Procena potro≈°nje goriva automobila prema broju cilindara.  
  - Prognoza temperature na osnovu vla≈ænosti vazduha.  

Linearna regresija poku≈°ava da pronaƒëe pravu liniju koja najbolje opisuje odnos izmeƒëu ulaznih i izlaznih podataka.

Matematiƒçki, jednostavna linearna regresija se izra≈æava kao:

ùë¶=ùë§ùëã+ùëè

gde su:    
w ‚Äì koeficijent (nagib linije)   
b ‚Äì slobodan ƒçlan (presek sa y-osom)


![img/linear_regression.gif](img/linear_regression.gif)

---

### üìö **Vrste regresije**  
| Tip                  | Opis                                  | Primer                      |  
|-----------------------|---------------------------------------|-----------------------------|  
| **Linearna**          | Prava linija koja modeluje vezu.      | `cena = 2000 √ó kvadratura + 50.000` |  
| **Polinomijalna**        | Kriva linija (npr. kvadratna, kubna). | `cena = 3√ókvadratura¬≤ + 1000√ókvadratura + 10.000` |  
| **Logistiƒçka**        | Predviƒëanje verovatnoƒáe (0-1).        | Verovatnoƒáa da je e-mail spam. |  

---

### üéØ **Kljuƒçni koncepti**  
#### 1. **Zavisna i nezavisna promenljiva**  
- **Nezavisna (X)**: Ulaz koji koristimo za predviƒëanje (npr. kvadratura).  
- **Zavisna (Y)**: Izlaz koji ≈æelimo da predvidimo (npr. cena).  
- **Primer**:  
  Ako trenirate da predvidite vreme, X su oblaci, Y je ki≈°a.  
  Napomena: Ulaza mo≈æe biti vi≈°e i tada govorimo o vi≈°estrukoj linearnoj regresiji.

#### 2. **Koeficijenti (Parametri)**  
- Vrednosti koji defini≈°u odnos izmeƒëu X i Y.  
- **Primer**: U `cena = 2000 √ó kvadratura + 50.000`:  
  - `2000` = uticaj kvadrature na cenu.  
  - `50.000` = osnovna cena (kad je kvadratura 0).  


#### 3. **Funkcija gubitka (Loss)**  
- **Najƒçe≈°ƒáe**: **Mean Squared Error (MSE)** ‚Äì proseƒçna kvadratna gre≈°ka.  
  ```  
  MSE = (1/n) * Œ£ (stvarno - predviƒëeno)¬≤  
  ```  
- **üåü Analogija**: Kazna za svako lo≈°e predviƒëanje ‚Äì ≈°to ste dalje, veƒáa kazna.  

---

### ‚öôÔ∏è **Kako radi jednostavna linearna regresija?**  
1. **Inicijalizacija**: Model poƒçinje sa nasumiƒçnim koeficijentima (npr. `cena = 100 √ó kvadratura + 0`).  
2. **Predviƒëanje**: Raƒçuna Y za sve X vrednosti.  
3. **Raƒçunanje gre≈°ke**: MSE pokazuje koliko je predviƒëanje lo≈°e.  
4. **Pode≈°avanje koeficijenata**: Algoritam (npr. gradijentni spust) menja koeficijente da smanji gre≈°ku.  
5. **Ponavlja se** dok gre≈°ka ne postane minimalna.

---

### üåç **Primer: Predviƒëanje cene stanova**  
1. **Podaci**:

| Kvadratura | Cena (‚Ç¨) |
|------------|----------|
| 50         | 150.000  |
| 80         | 220.000  |
| 100        | 280.000  |

2. **Model**: Linearna regresija nalazi pravac koji najbolje prati podatke.  
   - **Jednaƒçina**: `cena = 2000 √ó kvadratura + 50.000`  
   - **Tumaƒçenje**: Svaki dodatni m¬≤ poveƒáava cenu za 2000‚Ç¨, osnovna cena je 50.000‚Ç¨.  

3. **Predviƒëanje za 120 m¬≤**:  
   `2000 √ó 120 + 50.000 = 290.000‚Ç¨`.  

---

### üìä **Evaluacija modela**  
| **Metrika**            | **Formula**                                   | **Opis**                                |
|-----------------------|--------------------------------------------|-----------------------------------------|
| **MSE**               |  $MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$  | Velike gre≈°ke su stro≈æe ka≈ænjene jer se kvadriraju. |
| **R¬≤ (R-squared)**    | $R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$ | Pokazuje koliko dobro model obja≈°njava varijansu podataka (od 0 do 1). |
| **MAE**               | $MAE = \frac{1}{n} \sum_{i=1}^{n} \left\| y_i - \hat{y}_i \right\|$                 | Proseƒçna apsolutna gre≈°ka, lako za tumaƒçenje. |


- **Primer**:  
  - **R¬≤ = 0.85** ‚Üí Model obja≈°njava 85% varijanse u podacima.  
  - **MSE = 10.000** ‚Üí Proseƒçna kvadratna gre≈°ka je 10.000‚Ç¨¬≤.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

csv_url = "https://raw.githubusercontent.com/ywchiu/riii/master/data/house-prices.csv"
data = pd.read_csv(csv_url)
data.head()
````

````python
data.describe()
````

````python
data["m2"] = data["SqFt"]*0.09
data.describe()
````

Kvadratura stana je ulaz, cena stana je izlaz, ono ≈°to predviƒëamo

````python
X = data[['m2']].values
Y = data['Price'].values
````

Kreiramo model linearne regresije i metodom fit pronalazimo funkciju koja najbolje opisuje zavisnost

````python
model = LinearRegression()
model.fit(X, Y)
````

Mo≈æemo prikazati koeficijente dobijene posle treniranja

````python
intercept = model.intercept_
slope = model.coef_[0]

print(f"Presek (a): {intercept:.2f}")
print(f"Nagib (b): {slope:.2f}")
````

Hajde vizuelno da prika≈æemo:

````python

plt.scatter(X, Y, c='b', marker='o', label='Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel("m2")
plt.ylabel("House Price")
plt.title("Linear Regression: House Price vs. Area")
plt.legend()
plt.show()
````

## Vi≈°estruka linearna regresija

Atributi i ciljna promenljiva:

````python
cols = ['m2', 'Bedrooms']
X = data[cols].values
Y = data['Price'].values
````

Obuƒçavanje:

````python
model = LinearRegression()
model.fit(X, Y)
````

Parametri:

````python
intercept = model.intercept_
slope_1 = model.coef_[0]
slope_2 = model.coef_[1]

print(f"Presek (a): {intercept:.2f}")
print(f"Nagib za atribut m2 (b1): {slope_1:.2f}")
print(f"Nagib za atribut Bedrooms (b2): {slope_2:.2f}")
````

````python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# scatter podaci
ax.scatter(data['m2'], data['Bedrooms'], data['Price'], c='blue', marker='o', label='Data')

# mre≈æa za regresionu ravan
sqft_range = np.linspace(data['m2'].min(), data['m2'].max(), 30)
bedrooms_range = np.linspace(data['Bedrooms'].min(), data['Bedrooms'].max(), 30)
sqft_grid, bedrooms_grid = np.meshgrid(sqft_range, bedrooms_range)
X_grid = np.c_[sqft_grid.ravel(), bedrooms_grid.ravel()]
price_pred = model.predict(X_grid).reshape(sqft_grid.shape)
# povr≈°ina (ravan)
ax.plot_surface(sqft_grid, bedrooms_grid, price_pred, color='red', alpha=0.5)
ax.set_xlabel('m2')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('House Price')
ax.set_title('Linear Regression: Price vs m2 & Bedrooms')

plt.show()
````

Koeficijenti uz atribute nam ukazuju na bitnost tog atributa za ciljnu promenljivu:

````python
coefficients = pd.DataFrame(model.coef_, ['m2', 'Bedrooms'], columns=['Coefficient'])
print(coefficients)

plt.figure(figsize=(12, 6))
sns.barplot(x=coefficients['Coefficient'], y=coefficients.index)
plt.title('Feature Importance')
plt.xlabel('Coefficient Value')
plt.ylabel('Features')
plt.show()
````
